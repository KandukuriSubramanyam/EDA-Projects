# Outliers Detection
* via 'Boxplot' and 'IQR method by knowing upper_limit and lower_limit'

```python
plt.figure(figsize=(5,5))
plotnumber=1
for column in df:
    if plotnumber <= 4:
        plt.subplot(2,2,plotnumber)
        sns.boxplot(x = df[column])
        plt.xlabel(column,fontsize=10)
        plotnumber+=1
plt.tight_layout()
```

* Seems like SepalWidthCm column has outlier data points and I'll try to find them via IQR method by knowing upper_limit and lower_limit

```python
Q3 = df.SepalWidthCm.quantile(0.75)
```

```python
Q3
```

```python
Q1 = df.SepalWidthCm.quantile(0.25)
```

```python
Q1
```

```python
IQR = Q3 - Q1
```

```python
IQR
```

```python
upper_limit = Q3 + (1.5*IQR)
upper_limit
```

```python
lower_limit = Q1 - (1.5*IQR)
lower_limit
```

```python
df.loc[df.SepalWidthCm >= upper_limit,'SepalWidthCm']
```

* There are 3 outlier data points which are greater than upper_limit value

```python
df.loc[df.SepalWidthCm <= lower_limit,'SepalWidthCm']
```

* There is 1 outlier data point which is less than lower_limit value

```python
sns.displot(x=df.SepalWidthCm,kde=True)
```

```python
df.SepalWidthCm.skew()
```

```python
df.SepalWidthCm.kurtosis()
```

* SepalWidthCm is not a normally distributed data hence ouliers data needs to be replaced with median of SepalWidthCm column during preprocessing step.
