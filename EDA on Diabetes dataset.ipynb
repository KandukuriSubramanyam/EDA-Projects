## Bivariate Analysis

```python
plt.figure(figsize=(10,10))
plotnumber=1
for i in df:
    if plotnumber <= 8:
        plt.subplot(4,2,plotnumber)
        sns.regplot(data=df,x=i,y=df.Outcome)
        plt.xlabel(xlabel=i,fontsize=10)
        plotnumber+=1
plt.tight_layout()
```

Linear Relationships:
* Pregnancies vs Outcome: As the no. of pregnancies is increasing, a person being diabetic is also increasing and it is of moderate positive.
* Glucose vs Outcome: As the blood glucose level is increasing, a person being diabetic is also increasing and it is of strong positive.
* BloodPressure vs Outcome: This feature doesn't seem to be too much impactful on making a person diabetic and it is of weak positive.
* SkinThickness vs Outcome: This feature also have less impact on making a person diabetic and it is of weak positive.
* Insulin vs Outcome: As the levels of Insulin increases, a person being diabetic is decreasing but has few data points making diabetic even if the level of insulin producer is high and therefore, they are of moderate positive.
* BMI vs Outcome: As the value of BMI increases, the chance of being diabetic is also increasing and they are having moderate positive.
* DiabetesPedigreeFunction vs Outcome: As there is daibetes in the family history, there is an increasing chance of a person being diabetic. They have weak positive.
* Age vs Outcome: As the age of a person increasing, the chance of becoming diabetic is also increasing and they have a moderate positive.

```python
plt.figure(figsize=(10,10))
plotnumber = 1
coc = len(df.columns)
for i in range(coc-1):
    if plotnumber <= 8:
        plt.subplot(4,2,plotnumber)
        c1 = df.columns[i]
        c2 = df.columns[i+1]
        sns.regplot(data = df, x=c1, y=c2)
        plotnumber+= 1
plt.tight_layout()
```

# Outlier Detection
* via 'Boxplot' and 'IQR method by knowing upper_limit and lower_limit'


```python
plt.figure(figsize=(10,15))
plotnumber=1
for i in df:
    if plotnumber <= 9:
        plt.subplot(5,2,plotnumber)
        sns.boxplot(x=df[i])
        plotnumber+=1
plt.tight_layout()
```

* Seems like all Independent features have the outlier data points in it.

```python
Q3 = df.Pregnancies.quantile(0.75)
```

```python
Q3
```

```python
Q1 = df.Pregnancies.quantile(0.25)
Q1
```

```python
IQR = Q3 - Q1
IQR
```

```python
upper_limit = Q3 + (1.5*IQR)
upper_limit
```

```python
df.loc[(df.Pregnancies)>upper_limit,'Pregnancies']
```

* There are 4 extreme data points which are greater than upper limit in the Pregnancies data points.

```python
Q3 = df.Glucose.quantile(0.75)
```

```python
Q3
```

```python
Q1 = df.Glucose.quantile(0.25)
Q1
```

```python
IQR = Q3 - Q1
IQR
```

```python
lower_limit = Q1 - (1.5*IQR)
lower_limit
```

```python
df.loc[(df.Glucose)<lower_limit,'Glucose']
```

* There are 5 outlier data points less than lower limit in the Glucose data points.

```python
Q3 = df.BloodPressure.quantile(0.75)
Q3
```

```python
Q1 = df.BloodPressure.quantile(0.25)
Q1
```

```python
len(df.loc[(df.BloodPressure>upper_limit)|(df.BloodPressure<lower_limit),'BloodPressure'])
```
