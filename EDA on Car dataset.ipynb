## Bivariate Analysis

```python
plt.figure(figsize=(10,25))
plotnumber=1
for column in df:
    if plotnumber <= 11:
        plt.subplot(6,2,plotnumber)
        sns.scatterplot(x=df[column],y=df.mpg)
        plotnumber+=1
plt.tight_layout()
```

* cyl vs mpg: As the no of cylinders used in car increasing, the mileage decreases and is with weak negative linear relationship.
* disp vs mpg: As the engine displacement increases, the mileage decreases and is with moderate negative linear relationship.
* hp vs mpg: As the engine horse power increase, the mileage decreases and is with moderate negative linear relationship.
* drat vs mpg: As the drat value increases, the mileage also increases and is with moderate positive linear relationship.
* wt vs mpg: As the car weight increases, the mileage decreases and is with strong negative linear relationship.
* qsec vs mpg: Most of the times, as the qsec value is medium, the mileage is high and is not showing any linear relationship.
* vs vs mpg: The mileage is high for straight engine and mileage is low for v-shaped engine, the relationship is of weak positive linear relationship.
* am vs mpg: The mileage is high for manual transmission cars and less for automatic transmission cars, it is of weak positive linear relationship.
* gear vs mpg: The mileage is high for cars with more no of gears and it is of moderate positve linear relationship.
* carb vs mpg: The mileage is high when the no. of carburetors is less and it is of moderate negative linear relationship.

```python
plt.figure(figsize=(10,25))
plotnumber=1
loc = len(df.columns)
for i in range(loc-1):
    if plotnumber <= 10:
        plt.subplot(6,2,plotnumber)
        c1 = df.columns[i]
        c2 = df.columns[i+1]
        sns.regplot(data=df,x=c1,y=c2)
        plt.xlabel(df.columns[i],fontsize=15)
        plt.ylabel(df.columns[i+1],fontsize=15)
        plotnumber+=1
plt.tight_layout()
```

* mpg vs cyl: moderate negative linear relationship.
* cyl vs disp: moderate positive linear relationship.
* disp vs hp: moderate positive linear relationship.
* hp vs drat: weak negative linear relationship.
* drat vs wt: moderate negative linear relationship.
* wt vs qsec: weak negative linear relationship.
* qsec vs vs: weak positive linear relationship.
* vs vs am: no linear relationship.
* am vs gear: weak positive linear elationship.
* gear vs carb: weak positive linear relationship.

# Outliers Detection
* via 'Boxplot' and 'IQR method by knowing upper_limit and lower_limit'

```python
df.info()
```

```python
plt.figure(figsize=(5,5))
plotnumber=1
for column in df:
    if plotnumber <= 11:
        plt.subplot(6,2,plotnumber)
        sns.boxplot(x = df[column])
        plt.xlabel(column,fontsize=10)
        plotnumber+=1
plt.tight_layout()
```

* Seems like columns mpg,hp,wt,qsec and carb has datapoints to the right from the 100th percentile only and not on left. Hence, I'll try to find them via IQR method by knowing upper_limit only in this case.

## mpg

```python
Q3 = df.mpg.quantile(0.75)
```

```python
Q3
```

```python
Q1 = df.mpg.quantile(0.25)
```

```python
Q1
```

```python
IQR = Q3 - Q1
```

```python
IQR
```

```python
upper_limit = Q3 + (1.5*IQR)
upper_limit
```

```python
df.loc[df.mpg >= upper_limit,'mpg']
```

* There is 1 outlier data point which is greater than upper_limit value

Strong positove correlations:
* mpg (vs) drat indicates fuel efficiency increases in the increase of rear axle ratio.
* cyl (vs) wt, hp, disp indicates with the increase in no. of culinders, weight, horse power and displacement also increases.
* disp (vs) hp indicates that as displacement of the piston increases the horse power also increases.
* hp (vs) carb indicates that the horse power increases as the no. of carburetors increase.
* drat (vs) gear, am indicates that rear axle ration increases as the no. of gears increasing and it also increases for automatic transmission models.
